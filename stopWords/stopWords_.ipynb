```python
def preprocess(text):
    doc = nlp(text)
    stopword_olmayanlar = [token.text for token in doc if not token.is_stop]
    return stopword_olmayanlar
```

```python
preprocess("Dr Hamilton is waiting for a job")
```

```python
preprocess("Steven is eating pizza in the street")
```

```python
def p(text):
    doc = nlp(text)

    stopword_olmayanlar = [token.text for token in doc if not token.is_stop]
    return " ".join(stopword_olmayanlar) #liste olarak değil cümle olarak gelmesini sağlar
```

```python
p("Dr Hamilton is waiting for a job")
```
